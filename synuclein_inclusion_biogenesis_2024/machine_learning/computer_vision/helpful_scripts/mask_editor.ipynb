```python
import os
import glob
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import numpy as np
```

```python
def binarize_masks(input_folder, output_folder):
    # Create output folder if missing
    os.makedirs(output_folder, exist_ok=True)

    exts = ["*.png", "*.jpg", "*.jpeg", "*.tif", "*.tiff"]

    # Collect mask paths
    mask_paths = []
    for e in exts:
        mask_paths.extend(glob.glob(os.path.join(input_folder, e)))

    mask_paths = sorted(mask_paths)

    print(f"Found {len(mask_paths)} mask images.\n")

    if not mask_paths:
        print("No mask images found.")
        return

    for p in mask_paths:
        mask = imageio.imread(p)

        # Convert everything ≥1 → 1
        binary_mask = (mask >= 1).astype(np.uint8)

        # Save into output folder
        filename = os.path.basename(p)
        save_path = os.path.join(output_folder, filename)

        imageio.imwrite(save_path, binary_mask)

        print(f"Saved: {save_path}")

    print("\nAll masks binarized and saved!")
```

```python
if __name__ == "__main__":
    input_folder = "../images/new_inclusion_masks_neeharika"
    output_folder = "../images/inclusion_annotated_dataset"
    binarize_masks(input_folder, output_folder)
```
